```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/Users/juliaodintsova/Desktop/Python2/kc_house_data.csv')

plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=15);
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Кол-во');
```

Диапазон цен находится от 50000 до 8000000

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['sqft_living'], bins=15);
plt.title('Распределение жилой площади')
plt.xlabel('Площадь')
plt.ylabel('Кол-во');
```

Наибольшее количество домов с квадратурой жилой около 1000

```python
plt.figure(figsize=(6, 6))
sns.histplot(df['yr_built'], bins=15);
plt.title('Распределение домов по году постройки')
plt.xlabel('Год постройки')
plt.ylabel('Кол-во');
```

Наибольшее количество домов построили в 2007 году и в 1980

```python
plt.figure(figsize=(6,5))
sns.histplot(df['waterfront'], bins=15);
plt.title('Распределение наличия вида на набережную')
plt.xlabel('Наличие вида на набережную')
plt.ylabel('Кол-во');
```

```python
plt.figure(figsize=(6, 6))
sns.histplot(df['floors'], bins=15);
plt.title('Распределение домов по количеству этажей в доме')
plt.xlabel('Количество этажей')
plt.ylabel('Кол-во');
```

Большинство домов имеют 1 или 2 этажа
